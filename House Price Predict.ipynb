```python
#Train verisini de ikiye ayırdık. Hedef sütunumuz fiyat olduğu için bu sütunu çıkardık
x=df_train.drop(['SalePrice'],axis=1)
y=df_train['SalePrice']#Bu değişkene fiyat sütununu atadık
```

```python
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression, ElasticNet, Ridge, Lasso
from sklearn.tree import ExtraTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor

from sklearn.model_selection import train_test_split
```

```python
def algo_test(x, y):
    
    L = LinearRegression()
    E = ElasticNet()
    R = Ridge()
    Lass = Lasso()
    ETR = ExtraTreeRegressor()
    GBR = GradientBoostingRegressor()
    XGBC = XGBRegressor()
    
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state=42)
    
    algos = [L, E, R, Lass, ETR, GBR, XGBC]
    algo_names= ['Linear', 'ElasticNet', 'Ridge', 'Lasso', 'Extra Tree', 'Gradient Boosting', 'XGBRegressor']
    r_squared = []
    rmse = []
    mae = []
    
    result = pd.DataFrame(columns = ['R_Squared', 'RMSE', 'MAE'],
                           index = algo_names)
    
    for algo in algos:
        algo.fit(x_train, y_train)
        algo.predict(x_test)
        
        r_squared.append(r2_score(y_test, algo.predict(x_test)))
        rmse.append(mean_squared_error(y_test, algo.predict(x_test))**.5)
        mae.append(mean_absolute_error(y_test, algo.predict(x_test)))
        
    result.R_Squared = r_squared
    result.RMSE = rmse
    result.MAE = mae
    
    return result.sort_values('R_Squared', ascending=False)
```

```python
#yedi tane regresyon algoritması tanımlı algo_test fonsiyonuna verilerimizi verdik
#ve başarı oranına göre büyükten küçüğe sıraladık
algo_test(x, y)
```
